```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (9, 5))
plt.hist(df['MedHouseVal'], bins = 15,  color='green', rwidth = 0.9, density = True)
plt.title('Histogram of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (9, 5))
plt.scatter(df['Longitude'], df['Latitude'], c = df['MedHouseVal'], alpha = 0.8, marker = '*')
plt.colorbar(label = 'Median House Value')
plt.title('Longitude vs Latitude Colored by Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (9, 5))
sns.boxplot(x = df['IncomeCategory'], y = df['MedHouseVal'], color = 'purple')
plt.title('Boxplot of Median House Value Grouped by Income Categories')
plt.xlabel('Income Category')
plt.ylabel('Median House Value')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64'])
```

```python
correlation_matrix = numerical_columns.corr()
```

```python
plt.figure(figsize = (9, 5))
sns.heatmap(correlation_matrix, annot = True)
plt.xticks(rotation = 45)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
average_values = df.groupby('High_Population_Area')['MedHouseVal'].mean().reset_index()
plt.figure(figsize = (8, 5))
sns.barplot(x = 'High_Population_Area', y = 'MedHouseVal', data = average_values, color = 'yellow')
plt.title('Average Median House Value by Population Area')
plt.xlabel('High Population Area')
plt.ylabel('Average Median House Value')
plt.show()
```
